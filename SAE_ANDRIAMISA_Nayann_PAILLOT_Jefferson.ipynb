<p style="font-size:1.1em;">
    Le nombre d'or (ou phi φ) est un nombre qui a la particularité de pouvoir vérifier l'équation $$ x^2=x+1$$<br> et qui veut dire autrement que lorsqu'on met le nombre d'or au carré, ça revient à lui rajouter 1.
    <br>
    Lors de la résolution de cette équation on retrouve une valeur de phi qui est de $$ \phi = \frac{1 + \sqrt{5}}{2}$$
    <br>
    Pour trouver l'extremum d'une fonction, on utilisera cependant $$1-\phi \quad ou \quad \frac{\sqrt{5}-1}{2}$$ afin d'avoir les mêmes proportions pour n'importe quelle itération au cours de la recherche, ce qui est aussi une particularité du nombre d'or appelé la proportion d'or.
</p>

```python
# Nombre d'or
phi = (sqrt(5) - 1)/2
```

<div style="background-color : rgba(0, 255, 0, 0.57); border : 2px solid black; border-radius : 19px; margin : 0% 10%; padding : 1%; text-align : center; font-size: 2em;">
    Premier Exemple
</div>

<p style="font-size:1.1em;">
    Nous allons prendre comme premier exemple la fonction :
    $f(x) = x^2-x+\frac{1}{2}$
</p>

```python
## Fonction f(x) = x² - x + 1/2
def f(x):
    return x**2 - x + (1/2)
```

<p style="font-size:1.1em;margin:10px;">
    Pour trouver l'extremum local on doit déjà définir l'intervalle dans lequel nous allons chercher.
</p>

```python
### Définition des intervalles sur lequel la recherche de l'extremum sera faite
## [-0.4 ; 1.3]
minX1 = -0.4
maxX1 = 1.3


## Boucle pour lister tous les x dans l'intervalle séparé de 0.01 chacun
i = minX1
x = []
while i < maxX1:
    x.append(i)
    i = i + 0.01

        
## Boucle pour lister tous les y, ou f(x), dans l'intervalle
j = 0
y = []
while j < len(x):
    y.append(f(x[j]))
    j += 1
    
plot(x, y)                              ## Courbe de f(x) avec x comme abscisse et y comme ordonnée
title('f(x) = x² - x + 1/2')            ## Titre du graphique
show()                                  ## Affiche la courbe de la fonction f(x)
```

<p style="font-size:1.1em; margin-top:25px;">
    Et nous allons ensuite placer deux points ($x1$ et $x2$) sur cette courbe et qui vérifierons la propriété de la proprtion d'or, soit qu'un point aura la même distance $d$> entre une extremité de l'intervalle $a$ que l'autre point aura avec l'autre extrémité de l'intervalle $b$.
    <br>
    Soit $$d = \phi\times(b-a), \quad x1 = a + d \quad et \quad x2 = b - d$$
</p>

```python
d = phi * (maxX1 - minX1) # d est la proportion d'or qui sera utilisée pendant la recherche


x1 = minX1 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
x2 = maxX1 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui


plot(x, y)                                         ## Courbe de f(x) avec x comme abscisse et y comme ordonnée
plot(x1, f(x1), marker = "o", color = "red")       ## Point x1 en rouge sur la courbe 
plot(x2, f(x2), marker = "o", color = "green")     ## Point x2 en vert sur la courbe
title('f(x) = x² - x + 1/2')                       ## Titre du graphique
show()                                             ## Affiche la courbe de la fonction f(x)
```

<p style="font-size:1.1em; margin:10px;">
    On va maintenant réduire l'intervalle autour du minimum et en fonction des deux points <em><strong>x1</strong></em> et <em><strong>x2</strong></em>.
    <br>
    Tel que, 
    <br>
    si $ f(x1) < f(x2) $, le nouvel intervalle devient $[x2:b]$
    <br> 
    et 
    <br> si $ f(x1) > f(x2) $, le nouvel intervalle devient $[a:x1]$
    <br><br>
    Puis on fera exactement la même chose qu'aux précédentes étapes.
</p>

```python
if (f(x1) < f(x2)) :
    minX1 = x2       ## L'extrémité gauche devient x2
    maxX1 = maxX1    ## L'extrémité droite ne change pas


    ## Boucle pour lister tous les x dans l'intervalle séparé de 0.01 chacun
    i = minX1
    x = []
    while i < maxX1:
        x.append(i)
        i = i + 0.01


    ## Boucle pour lister tous les y, ou f(x), dans l'intervalle
    j = 0
    y = []
    while j < len(x):
        y.append(f(x[j]))
        j += 1
    d = phi * (maxX1 - minX1) # d est la proportion d'or qui sera utilisée pendant la recherche

    x1 = minX1 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
    x2 = maxX1 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui


    plot(x, y)                                         ## Courbe de f(x) avec x comme abscisse et y comme ordonnée
    plot(x1, f(x1), marker = "o", color = "red")       ## Point x1 en rouge sur la courbe 
    plot(x2, f(x2), marker = "o", color = "green")     ## Point x2 en vert sur la courbe
    title('f(x) = x² - x + 1/2')                       ## Titre du graphique
    show()                                             ## Affiche la courbe de la fonction f(x) dans le nouvel intervalle
    
else :
    maxX1 = x1       ## L'extrémité droite devient x1
    minX1 = minX1    ## L'extrémité gauche ne change pas


    ## Boucle pour lister tous les x dans l'intervalle séparé de 0.01 chacun
    i = minX1
    x = []
    while i < maxX1:
        x.append(i)
        i = i + 0.01


    ## Boucle pour lister tous les y, ou f(x), dans l'intervalle
    j = 0
    y = []
    while j < len(x):
        y.append(f(x[j]))
        j += 1
        
    d = phi * (maxX1 - minX1) # d est la proportion d'or qui sera utilisée pendant la recherche

    x1 = minX1 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
    x2 = maxX1 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui


    plot(x, y)                                         ## Courbe de f(x) avec x comme abscisse et y comme ordonnée
    plot(x1, f(x1), marker = "o", color = "red")       ## Point x1 en rouge sur la courbe 
    plot(x2, f(x2), marker = "o", color = "green")     ## Point x2 en vert sur la courbe
    title('f(x) = x² - x + 1/2')                       ## Titre du graphique
    show()                                             ## Affiche la courbe de la fonction f(x) dans le nouvel intervalle
```

<p style="font-size:1.1em; margin:10px;">
    Et maintenant pour trouver une valeur approchée du minimum de la fonction, on répetera cela un certain nombre de fois et on fera la moyenne des x1 et des x2, pour après faire la moyenne de ces deux moyennes et obtenir une valeur approchée de l'extremum de la fonction
</p>

<p style="font-size: 1.3em; margin: 10px; color:red">
    Attention ! L'éxecution des cellules doivent se faire dans l'ordre pour que la cellule ci-dessous affiche correctement le minimum de la fonction.
    <br>
    Mais nous mettrons quand même une cellule qui regroupera toutes les étapes du début à la fin.
</p>

```python
moyenne_x1 = 0
moyenne_x2 = 0

######################################################

###  DEBUT DE LA FONCTION DE RECHERCHE DU MINIMUM  ###

######################################################

iteration = 0
while (iteration < 25) : ## On fera 25 itérations pour trouver une valeur approchée du minimum

## Boucle pour lister tous les x dans l'intervalle séparé de 0.01 chacun
    i = minX1
    x = []
    while i < maxX1:
        x.append(i)
        i = i + 0.01

        
## Boucle pour lister tous les y ou f(x) dans l'intervalle
    j = 0
    y = []
    while j < len(x):
        y.append(f(x[j]))
        j += 1
        
## d c'est la proportion d'or qui sera utilisée pendant la recherche
    d = phi * (maxX1 - minX1)
        
## x1 et x2 sont des points choisis sur l'intervalle selon la proportion d'or 
    x1 = minX1 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
    x2 = maxX1 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui
    
    
## Condition pour que la boucle ne fait que la représentation graphique des 5 premières itération
    if (iteration < 5): 
        plot(x, y)                                        ## Courbe de f(x) avec x comme abscisse et y comme ordonnée
        plot(x1, f(x1), marker = "o", color = "red")      ## Point x1 en rouge sur la courbe 
        plot(x2, f(x2), marker = "o", color = "green")## Point x2 en vert sur la courbe
        title('f(x) = x² - x + 1/2')                      ## Titre du graphique 
        show()                                            ## Affiche la courbe
    
    
## Condition de rétrécissement de l'intervalle en changeant soit l'extrémité droite soit l'extrémité gauche    
    if (f(x1) < f(x2)) :
        minX1 = x2     ## L'extrémité gauche devient x2
    else :
        maxX1 = x1     ## L'extrémité droite devient x1

        
    moyenne_x1 += x1   ## Somme des x1
    moyenne_x2 += x2   ## Somme des x2
        
    iteration += 1

moy_x1 = round(moyenne_x1/iteration, 2)   ## Calcul de la moyenne des x1
moy_x2 = round(moyenne_x2/iteration, 2)   ## Calcul de la moyenne des x2

minimum = round((moy_x1+moy_x2)/2,1)      ## Calcul de la moyenne des moyenne afin d'obtenir la valeur approchée du minimum

###################################################################################

###  AFFICHAGE DU RESULTAT DU MINIMUM DE f(x) AVEC LA MOYENNE DES x1 ET DES x2  ###

###################################################################################
print("Minimum de la fonction ≃ ", minimum) 
```

<p style="font-size:1.1em;">
    Nous allons prendre comme premier exemple la fonction :
    $g(x) = -2cos(x)^3 + sin(x) + \frac{3}{2}$
</p>

```python
## Fonction g(x) = -2cos(x)³ + sin(x) + 3/2
def g(x):
    return -2*cos(x)**3 + sin(x) +(3/2)

### Définition des intervalles sur lequel la recherche de l'extremum sera faite
## [-2 ; 0.25]
minX2 = -2
maxX2 = 0.25


moyenne_x1 = 0
moyenne_x2 = 0

######################################################
###  DEBUT DE LA FONCTION DE RECHERCHE DU MINIMUM  ###
######################################################

iteration = 0
while (iteration < 25) :  ## La fonction fera 25 itérations pour trouver une valeur approchée du minimum
    
    
## d c'est la proportion d'or qui sera utilisée pendant la recherche    
    d = phi * (maxX2 - minX2)
    

## Boucle pour lister tous les x dans l'intervalle séparé de 0.001 chacun
    i = minX2
    x = []
    while i < maxX2:
        x.append(i)
        i = i + 0.001
        
        
## Boucle pour lister tous les y ou f(x) dans l'intervalle
    j = 0
    y = []
    while j < len(x):
        y.append(g(x[j]))
        j += 1
        
        
## x1 et x2 sont des points choisis sur l'intervalle selon la proportion d'or 
    x1 = minX2 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
    x2 = maxX2 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui
    
    
## Condition pour que la boucle ne fait que la représentation graphique des 5 premières itération
    if (iteration < 5):
        plot(x, y)                                           ## Courbe de g(x) avec x comme abscisse et y comme ordonnée
        plot(x1, g(x1), marker = "o", color = "red")         ## Point x1 en rouge sur la courbe 
        plot(x2, g(x2), marker = "o", color = "green")       ## Point x2 en vert sur la courbe
        title('g(x) = -2cos(x)³ + sin(x) + 3/2')             ## Titre du graphique
        show()                                               ## Affiche la courbe
    
    
## Condition de rétrécissement de l'intervalle en changeant soit l'extrémité droite soit l'extrémité gauche    
    if (g(x1) < g(x2)) :
        minX2 = x2     ## L'extrémité gauche devient x2
    else :
        maxX2 = x1     ## L'extrémité droite devient x1

        
    moyenne_x1 += x1   ## Somme des x1
    moyenne_x2 += x2   ## Somme des x2
        
    iteration += 1

moy_x1 = round(moyenne_x1/iteration, 2)   ## Calcul de la moyenne des x1
moy_x2 = round(moyenne_x2/iteration, 2)   ## Calcul de la moyenne des x2

minimum = round((moy_x1+moy_x2)/2,1)      ## Calcul de la moyenne des moyenne afin d'obtenir la valeur approchée du minimum

###################################################################################

###  AFFICHAGE DU RESULTAT DU MINIMUM DE f(x) AVEC LA MOYENNE DES x1 ET DES x2  ###

###################################################################################

print("Minimum de la fonction ≃ ", minimum) 
```

<p style="font-size:1.1em;">
    Nous allons prendre comme premier exemple la fonction :
    $ h(x) = (\frac{\sqrt{5}}{2})x^2 -1 $
</p>

```python
# Fonction h(x) = (√5/2)x^2 - 1
def h(x):
    return (((sqrt(5))/2)*x**2 - 1)


### Définition des intervalles sur lequel la recherche de l'extremum sera faite
## [-1 ; 1]
minX3 = -1
maxX3 = 1


moyenne_x1 = 0
moyenne_x2 = 0


######################################################
###  DEBUT DE LA FONCTION DE RECHERCHE DU MINIMUM  ###
######################################################

iteration = 0
while (iteration < 20) :      ## La fonction fera 20 itérations pour trouver une valeur approchée du minimum
    

## d c'est la proportion d'or qui sera utilisée pendant la recherche
    d = phi * (maxX3 - minX3)

    
## Boucle pour lister tous les x dans l'intervalle séparé de 0.001 chacun
    i = minX3
    x = []
    while i < maxX3:
        x.append(i)
        i = i + 0.001

        
## Boucle pour lister tous les y ou f(x) dans l'intervalle       
    j = 0
    y = []
    while j < len(x):
        y.append(h(x[j]))
        j += 1
        
## x1 et x2 sont des points choisis sur l'intervalle selon la proportion d'or 
    x1 = minX3 + d      # x1 est le point qui a une distance de d entre l'extrémité à droite de l'intervalle et lui
    x2 = maxX3 - d      # x2 est le point qui a une distance de d entre l'extrémité à gauche de l'intervalle et lui


## Condition pour que la boucle ne fait que la représentation graphique des 5 premières itération
    if (iteration < 5):
        plot(x, y)                                          ## Courbe de h(x) avec x comme abscisse et y comme ordonnée
        plot(x1, h(x1), marker = "o", color = "red")        ## Point x1 en rouge sur la courbe 
        plot(x2, h(x2), marker = "o", color = "green")      ## Point x2 en vert sur la courbe
        title('h(x) = (√5/2)x^2 - 1')                       ## Titre du graphique
        show()                                              ## Affiche la courbe
    
    
## Condition de rétrécissement de l'intervalle en changeant soit l'extrémité droite soit l'extrémité gauche    
    if (h(x1) < h(x2)) :
        minX3 = x2     ## L'extrémité gauche devient x2
    else :
        maxX3 = x1     ## L'extrémité droite devient x1

        
    moyenne_x1 += x1   ## Somme des x1
    moyenne_x2 += x2   ## Somme des x2
        
    iteration += 1

moy_x1 = round(moyenne_x1/iteration, 2)   ## Calcul de la moyenne des x1
moy_x2 = round(moyenne_x2/iteration, 2)   ## Calcul de la moyenne des x2

minimum = round((moy_x1+moy_x2)/2,1)      ## Calcul de la moyenne des moyenne afin d'obtenir la valeur approchée du minimum

###################################################################################

###  AFFICHAGE DU RESULTAT DU MINIMUM DE f(x) AVEC LA MOYENNE DES x1 ET DES x2  ###

###################################################################################

print("Minimum de la fonction ≃ ", minimum) 
```
